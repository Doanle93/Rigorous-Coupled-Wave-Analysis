## Slanted Gratings
A slanted grating is kind of weird because you cannot do a Fourier decomposition along the x (or transverse) direction. The grating vector is now in some linear combination of x and z., which means something has to change, but not sure what.

Preliminarily, we can consider the simpliest slanted granting:
\begin{equation}
    \epsilon(x,z) = \epsilon_0+\Delta\epsilon \cos(K(x \cos(\phi) + z\sin(\phi)))
\end{equation}
which would require no Fourier decomposition(at least along the grating vector direction).

The question then is how do we consider the plane waves. We could still write incident waves in the Cartesian basis, but what happens when they hit the grating region?

One thought is, in the grating region:
\begin{equation}
    E = \sum_{i}U_xi + \exp[j(k_{xi}x+ k_{zi}z + iK_y y]
\end{equation}
Ordinarily, the decomposition would just contain the x and z components (the transverse components), but now we need to include the K_y component of the grating vector? This thought appears to be supported by the Moharam paper.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,1,1000)
z = np.linspace(-0.2, 0.2, 1000)
K =20
phi = 45*np.pi/180;
print(np.sin(phi))
[X,Z] = np.meshgrid(x,z);
eps = 2+1*np.cos(K*(X*np.cos(phi)+Z*np.sin(phi)))
```

```python
plt.figure();
plt.pcolormesh(X,Z,eps);
plt.clim(0,2.2)
plt.colorbar();
```
